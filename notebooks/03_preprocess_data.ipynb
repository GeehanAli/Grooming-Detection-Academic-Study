```python
import pandas as pd

# Load raw data
raw_path = 'data/raw/synthetic_chats_raw.csv'
df = pd.read_csv(raw_path)

print(df.shape)
df.head()
```

```python
import re

def clean_text(text):
    if pd.isna(text):
        return ""
    # lowercase
    text = text.lower()
    # remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['text_clean'] = df['text'].apply(clean_text)

df[['text', 'text_clean']].head()
```

```python
df = df.dropna(subset=['text_clean'])
print(df.shape)
```

```python
import os

os.makedirs('data/processed', exist_ok=True)

processed_path = 'data/processed/synthetic_chats_model_ready.csv'
df.to_csv(processed_path, index=False)

print("Processed dataset saved to:", processed_path)
```
